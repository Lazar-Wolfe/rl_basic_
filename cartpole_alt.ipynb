# Adding a callback to training stage

```python
import os
import gym 
from stable_baselines3 import PPO # proximal policy optimization
from stable_baselines3.common.vec_env import DummyVecEnv # DummyVecEnv is a vectorized environment
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
environemt_name = 'CartPole-v0'
env = gym.make(environemt_name)
env = DummyVecEnv([lambda: env]) # creating a vectorized environment
```

```python
save_path = os.path.join('Training','Saved_Models')
```

```python
log_path = os.path.join('Training','Logs')
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200,verbose = 1)
eval_callback = EvalCallback(env,callback_on_new_best=stop_callback,eval_freq=10000,best_model_save_path=save_path,verbose=1)
```

```python
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```

# Changing Policy


```python
net_arch=[dict(pi=[128, 128, 128, 128], vf=[128, 128, 128, 128])]
# specifyin a differnet architecture for the different neural networks used in PPO
#pi custom actor with 128 nodes
#vf value function with 128 nodes
```

```python
model = PPO('MlpPolicy', env, verbose = 1, policy_kwargs={'net_arch': net_arch})
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```
